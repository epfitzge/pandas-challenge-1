## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

df['Subtotal'] = df['unit_price'] * df['qty']

df
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

df['Shipping Weight'] = df['unit_weight'] * df['qty']

def shipping_price(weight):
    if weight > 50:
        return weight * 7
    else:
        return weight * 10

df['Shipping Price'] = df['Shipping Weight'].apply(shipping_price)

df
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

sales_tax = .0925


df['Total Price'] = ((df['Subtotal'] + df['Shipping Price']) * sales_tax) + df['Subtotal'] + df['Shipping Price']

df
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

df['Cost'] = (df['unit_cost'] * df['qty']) + df['Shipping Price']

df
```

```python
# Create a column for the profit of
# each line using line cost and line price

df['Total Profit'] = df['Total Price'] - df['Cost']

df
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.

top_id_one = 33615

top_id_one_spend = df.loc[df['client_id'] == top_id_one, 'Total Price'].sum()

top_id_two = 66037

top_id_two_spend = df.loc[df['client_id'] == top_id_two, 'Total Price'].sum()

top_id_three = 46820

top_id_three_spend = df.loc[df['client_id'] == top_id_three, 'Total Price'].sum()

top_id_four = 38378

top_id_four_spend = df.loc[df['client_id'] == top_id_four, 'Total Price'].sum()

top_id_five = 24741

top_id_five_spend = df.loc[df['client_id'] == top_id_five, 'Total Price'].sum()

print(f"The ID with the highest quantity spent the following: ${top_id_one_spend}")
print(f"The ID with the 2nd highest quantity spent the following: ${top_id_two_spend}")
print(f"The ID with the 3rd highest quantity spent the following: ${top_id_three_spend}")
print(f"The ID with the 4th highest quantity spent the following: ${top_id_four_spend}")
print(f"The ID with the 5th highest quantity spent the following: ${top_id_five_spend}")
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

filtered_totals_df = df.loc[df['client_id'].isin([top_id_one, top_id_two, top_id_three, top_id_four, top_id_five])]

summary_df = filtered_totals_df.groupby('client_id').agg({'qty': 'sum', 'Shipping Price': 'sum', 'Total Price': 'sum', 'Total Profit': 'sum'})

summary_df.sort_values(by='Total Profit', ascending = False)

```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

summary_df = summary_df.rename(columns={'qty': 'Quantity'})

summary_df['Formatted Quantity'] = summary_df['Quantity'].apply(lambda x: '{:,}'.format(x))

summary_df['Formatted Shipping Price ($M)'] = (summary_df['Shipping Price'] / 1_000_000).map('${:,.2f}'.format)

summary_df['Formatted Total Price ($M)'] = (summary_df['Total Price'] / 1_000_000).map('${:,.2f}'.format)

summary_df['Formatted Total Profit ($M)'] = (summary_df['Total Profit'] / 1_000_000).map('${:,.2f}'.format)

summary_df
```

```python
# Sort the updated data by "Total Profit" form highest to lowest

summary_df.sort_values(by='Formatted Total Profit ($M)', ascending = False)
```
